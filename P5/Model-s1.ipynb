### Baseline Solution

The baseline solution would be to predict the average sales of items by shops during November 2013 and November 2014.

```python
sales.head()
```

```python
nov2013 = sales[(sales['date'].dt.month == 11) & (sales['date'].dt.year == 2013)]
nov2014 = sales[(sales['date'].dt.month == 11) & (sales['date'].dt.year == 2014)]
```

```python
sample = pd.read_csv("data/sample_submission.csv")
sample.head()
```

```python
test.head()
```

```python
nov2013g = nov2013.groupby(["shop_id","item_id"])["item_cnt_day"].sum()
nov2014g = nov2014.groupby(["shop_id","item_id"])["item_cnt_day"].sum()
```

```python
nov2013g[5]
```

```python
submit = pd.DataFrame()
submit['ID'] = test['ID']
submit['item_cnt_month'] = 0
```

```python
submit
```

```python
ans = []
```

```python
for i in test['ID']:
    s = test.loc[i]['shop_id']
    j = test.loc[i]['item_id']
    try:
        s13 = nov2013g[s][j]
    except:
        s13 = 0
    try:
        s14 = nov2014g[s][j]
    except:
        s14 = 0
    ans.append((s13+s14)/2)
```

```python
submit['ID'] = test['ID']
```

```python
submit['item_cnt_month'] = ans
```

```python
submit
```

```python
submit.to_csv('submit.csv')
```
